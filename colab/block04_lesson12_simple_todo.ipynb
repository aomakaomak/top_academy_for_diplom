# Блок 4 — Матрицы

Матрица — это таблица чисел.
В машинном обучении матрица часто описывает данные: строки — объекты, столбцы — признаки.
На уроке:
1. создаём матрицу
2. смотрим её размер
3. создаём вектор
4. умножаем матрицу на вектор
5. делаем вывод

Формат урока: **5 кодовых ячеек**

Представьте, что вы хотите создать простую модель для предсказания цены дома. У каждого дома есть несколько характеристик, например, площадь (в квадратных метрах) и количество комнат. У вас есть данные по нескольким домам.

Создаём матрицу:

В машинном обучении матрица (например, X из вашего ноутбука) часто используется для хранения набора данных. Каждая строка матрицы представляет собой отдельный объект (в нашем случае, один дом), а каждый столбец представляет собой признак или характеристику этого объекта (например, площадь, количество комнат).
Допустим, у вас есть 3 дома:
Дом 1: Площадь = 100 м², Комнат = 3
Дом 2: Площадь = 150 м², Комнат = 4
Дом 3: Площадь = 200 м², Комнат = 5
В виде матрицы это будет выглядеть так:
X = [[100, 3],
     [150, 4],
     [200, 5]]
Смотрим её размер:

Свойство .shape (как X.shape в ноутбуке) показывает вам размерность матрицы, то есть количество строк и столбцов. Для нашей матрицы X.shape будет (3, 2). Это означает, что у нас есть 3 объекта (дома) и 2 признака для каждого объекта.
Это важно, чтобы убедиться, что ваши данные имеют правильную структуру для дальнейших вычислений.
Создаём вектор:

Вектор (например, w из ноутбука) часто представляет собой веса или коэффициенты в математической модели. Эти веса определяют, насколько важен каждый признак для предсказания целевой переменной (в нашем случае, цены).
Допустим, наша простая модель предсказывает цену как Цена = (вес_площади * Площадь) + (вес_комнат * Количество_комнат).
Мы можем задать вектор весов, например:
w = [10, 5]  # где 10 — вес для площади, 5 — вес для количества комнат
Умножаем матрицу на вектор:

Операция матричного умножения (X @ w в NumPy) позволяет одновременно применить нашу модель ко всем объектам в наборе данных.
По сути, для каждой строки матрицы X (каждого дома) происходит умножение значений признаков этого дома на соответствующие веса из вектора w, а затем суммирование результатов. Это дает нам предсказанную цену для каждого дома.
Например, для Дома 1: (100 * 10) + (3 * 5) = 1000 + 15 = 1015
Для Дома 2: (150 * 10) + (4 * 5) = 1500 + 20 = 1520
Для Дома 3: (200 * 10) + (5 * 5) = 2000 + 25 = 2025
Результатом будет новый вектор [1015, 1520, 2025], который содержит предсказанные цены для каждого из трех домов.
Вывод: Использование матриц и векторов, а также операция матричного умножения, является фундаментальным для эффективности в машинном обучении. Это позволяет нам выполнять одни и те же вычисления для большого количества объектов одновременно, значительно ускоряя процесс обучения и предсказания моделей.

## Ячейка 1 — создаём матрицу

Матрица — это просто двумерный массив чисел.

```python
import numpy as np

X = np.array([
    [1, 2],
    [3, 4],
    [5, 6]
])

print(X)
assert X.shape == (3, 2)
```

## Ячейка 2 — размер матрицы

Свойство `.shape` показывает число строк и столбцов.

```python
rows, cols = X.shape
print("Строк:", rows)
print("Столбцов:", cols)

assert rows == 3
assert cols == 2
```

## Ячейка 3 — создаём вектор

Этот вектор будет умножаться на матрицу.

```python
w = np.array([10, 1])
print("Вектор w:", w)

assert len(w) == 2
```

## Ячейка 4 — умножение

Используем оператор `@`. Это матричное умножение.

Формула матричного умножения (матрица на вектор)
Когда вы умножаете матрицу на вектор, результат представляет собой новый вектор. Каждый элемент этого результирующего вектора получается путем "скалярного произведения" соответствующей строки матрицы на вектор. То есть, вы умножаете каждый элемент строки матрицы на соответствующий элемент вектора и суммируете эти произведения.

Формально, если у нас есть матрица X размера m x n (m строк, n столбцов) и вектор w размера n (n элементов), то результат y будет вектором размера m. Каждый элемент y_i вычисляется как: y_i = X_i1 * w_1 + X_i2 * w_2 + ... + X_in * w_n Где X_ij — это элемент матрицы X в i-й строке и j-м столбце, а w_j — это j-й элемент вектора w.

Расчет результата в Ячейке 4
Применяем эту формулу к нашим X и w:

1. Для первой строки матрицы X ([1, 2]):

Берем первый элемент строки (1) и умножаем на первый элемент вектора w (10): 1 * 10 = 10
Берем второй элемент строки (2) и умножаем на второй элемент вектора w (1): 2 * 1 = 2
Суммируем эти произведения: 10 + 2 = 12 Это первый элемент результирующего вектора.
2. Для второй строки матрицы X ([3, 4]):

Берем первый элемент строки (3) и умножаем на первый элемент вектора w (10): 3 * 10 = 30
Берем второй элемент строки (4) и умножаем на второй элемент вектора w (1): 4 * 1 = 4
Суммируем эти произведения: 30 + 4 = 34 Это второй элемент результирующего вектора.
3. Для третьей строки матрицы X ([5, 6]):

Берем первый элемент строки (5) и умножаем на первый элемент вектора w (10): 5 * 10 = 50
Берем второй элемент строки (6) и умножаем на второй элемент вектора w (1): 6 * 1 = 6
Суммируем эти произведения: 50 + 6 = 56 Это третий элемент результирующего вектора.
Итоговый результат
Таким образом, результирующий вектор result после умножения X @ w равен [12, 34, 56], что полностью соответствует выводу в Ячейке 4.

Эта операция лежит в основе многих алгоритмов машинного обучения, например, для вычисления предсказаний в линейной регрессии или применения фильтров в сверточных нейронных сетях.

```python
result = X @ w
print("Результат умножения:", result)

assert len(result) == 3
```

## Ячейка 5 — вывод

Матрица помогает сразу обрабатывать много объектов.

Простыми словами, проверка говорит: "Я ожидаю, что result будет массивом NumPy. Если это не так, значит что-то пошло не так, и нужно остановить выполнение программы, чтобы разобраться."

```python
print("Каждая строка матрицы умножилась на вектор.")
print("Так в ML часто считают значения для всех объектов сразу.")

assert isinstance(result, np.ndarray)
```
