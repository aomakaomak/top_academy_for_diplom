# Блок 4 — Занятие  — Векторы

Вектор можно понимать как короткий список чисел.
В AI и ML векторы очень важны: они описывают объекты, пользователей, тексты и изображения.
На уроке:
1. создаём два вектора
2. смотрим их длину
3. складываем векторы
4. считаем скалярное произведение
5. рисуем векторы на плоскости

Формат урока: **5 кодовых ячеек**, одна ячейка = одна простая идея.

## Ячейка 1 — создаём два вектора

Используем `np.array(...)`. Это просто удобный способ хранить числа.

давайте подробнее разберем np.array([2, 1]):

np.array(...): Это функция из библиотеки NumPy (которую мы импортировали как np). Она используется для создания массивов (arrays). Массивы NumPy очень похожи на списки в Python, но они оптимизированы для выполнения математических операций, особенно с большими наборами числовых данных.

[2, 1]: Это обычный список Python, содержащий два числа: 2 и 1.

Когда вы пишете np.array([2, 1]), NumPy берет этот список [2, 1] и превращает его в свой собственный объект — одномерный массив (который в контексте линейной алгебры мы называем вектором). Этот вектор содержит элементы 2 и 1. В нашем примере v1 = np.array([2, 1]) создает вектор v1 с этими значениями.



```python
import numpy as np

v1 = np.array([2, 1])
v2 = np.array([1, 3])

print("v1 =", v1)
print("v2 =", v2)

assert len(v1) == len(v2) == 2
```

## Ячейка 2 — длина вектора

Длина вектора считается через `np.linalg.norm(...)`.

давайте разберем np.linalg.norm(v1):

np.linalg.norm(...): Это функция из подмодуля linalg (линейная алгебра) библиотеки NumPy. Она используется для вычисления нормы вектора или матрицы. В контексте векторов, norm обычно означает евклидову норму, которая является длиной вектора.

Когда вы применяете np.linalg.norm(v1), где v1 — это наш вектор np.array([2, 1]), эта функция вычисляет его длину. Длина вектора в двумерном пространстве рассчитывается как корень квадратный из суммы квадратов его координат (по теореме Пифагора). В данном случае это будет sqrt(2^2 + 1^2) = sqrt(4 + 1) = sqrt(5).

Таким образом, np.linalg.norm(v1) вернет численное значение длины вектора v1, которое составляет примерно 2.23

```python
len_v1 = np.linalg.norm(v1)
len_v2 = np.linalg.norm(v2)

print("Длина v1:", len_v1)
print("Длина v2:", len_v2)

assert len_v1 > 0 and len_v2 > 0
```

## Ячейка 3 — сложение векторов

Векторы одинаковой длины можно складывать поэлементно.

```python
v_sum = v1 + v2
print("Сумма векторов:", v_sum)

assert all(v_sum == np.array([3, 4]))
```

## Ячейка 4 — скалярное произведение

Используем `np.dot(...)`. Это очень важная операция в линейной алгебре и ML.

Формула скалярного произведения для двух векторов A = [a1, a2] и B = [b1, b2]: A · B = a1*b1 + a2*b2

Применяя эту формулу к нашим векторам:

v1 = np.array([2, 1])
v2 = np.array([1, 3])
Расчет: v1 · v2 = (2 * 1) + (1 * 3) v1 · v2 = 2 + 3 v1 · v2 = 5

```python
dot_value = np.dot(v1, v2)
print("Скалярное произведение:", dot_value)

assert dot_value == 5
```

## Ячейка 5 — рисунок

Рисуем векторы как стрелки на плоскости.

```python
import matplotlib.pyplot as plt

plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1)
plt.xlim(0, 4)
plt.ylim(0, 4)
plt.title("Два вектора")
plt.grid()
plt.show()
```
