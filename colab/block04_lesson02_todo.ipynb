## Мини-теория

### Распределение
Показывает, как часто встречаются значения в данных.

### Нормальное распределение
- Симметричное
- Среднее ≈ медиана
- 68% значений лежат в диапазоне ±1 std

### Гистограмма
Строится через matplotlib:
- plt.hist()
- plt.show()

В конце: Run all → все assert должны пройти.


## 1/10 Генерация нормального распределения

```python

import numpy as np

data = np.random.normal(0, 3, 1000)

len(data)
```

## 2/10 Среднее и стандартное отклонение

```python
mean = np.mean(data)
std = np.std(data)
mean, std
```

## 3/10 Гистограмма нормального распределения

```python
import matplotlib.pyplot as plt
plt.hist(data, bins=30)
plt.title("Normal Distribution")
plt.show()
```

## 4/10 Добавление выброса

```python
data_out = np.append(data, 10)
len(data_out)
```

## 5/10 Гистограмма с выбросом

```python
plt.hist(data_out, bins=30)
plt.title("With Outlier")
plt.show()
```

## 6/10 Медиана

```python
median = np.median(data_out)
median
```

## 7/10 Проверка правила 68%

```python
lower = mean - std
upper = mean + std
within = np.sum((data >= lower) & (data <= upper))
within / len(data)
```

## 8/10 Равномерное распределение

```python
uniform = np.random.uniform(-6, 6, 1000)
plt.hist(uniform, bins=30)
plt.title("Uniform Distribution")
plt.show()
```

## 9/10 Сравнение средних

```python
np.mean(data), np.mean(uniform)
```
